```python
import glob
import cv2
from PIL import Image
import numpy as np
```

```python
def imshow(a):
    a = a.clip(0, 255).astype('uint8')
    if a.ndim == 3:
        if a.shape[2] == 4:
            a = cv2.cvtColor(a, cv2.COLOR_BGRA2RGBA)
        else:
            a = cv2.cvtColor(a, cv2.COLOR_BGR2RGB)
    display(Image.fromarray(a))
```

```python
files = glob.glob("data/cards_jpeg/*.jpeg")
```

```python
images = [cv2.imread(f, 1) for f in files]
```

```python
cards = [cv2.resize(img, None, fx=0.1, fy=0.1) for img in images]
imshow(cards[0])
```

```python
im = cards[0].copy()
imshow(im)
imgray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
imshow(imgray)
thresh = (imgray > 180).astype('uint8')
imshow(thresh*255)

contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
areas = [cv2.contourArea(c) for c in contours]
max_area = max(areas)
for i in range(len(contours)):
    if np.abs(areas[i] - max_area) < 20:
        cnt = contours[i]
        break
cv2.drawContours(im, [cnt], -1, (0,255,0), 3)
imshow(im)
```
